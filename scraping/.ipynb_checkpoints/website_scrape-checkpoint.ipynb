```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup

import time

from datetime import datetime, timedelta
from dateutil import parser

import pandas as pd
```

```python
# sample input:
# '2 hours ago ... Bombs went off and lights went out as a blackout plunged Gaza into chaos. Israeli forces have begun what Prime Minister Benjamin Netanyahu\xa0...'
# from
# div_list[0].find('div', class_='gs-bidi-start-align gs-snippet').text


def convert_to_local_time(time_string):
    
    try:
        # time_string = soup_find_string.split("...")[0].rstrip()
        time_strings_list = time_string.split()
        duration_str = time_strings_list[1]

        if (time_string == 'just now'):
            time_stamp = datetime.now()
        elif (duration_str == 'second') or (duration_str == 'seconds'):
            delta_num = int(time_strings_list[0])
            time_stamp = datetime.now() - timedelta(seconds=delta_num)
        elif (duration_str == 'minute') or (duration_str == 'minutes') or (duration_str == 'mins') or (duration_str == 'min'):
            delta_num = int(time_strings_list[0])
            time_stamp = datetime.now() - timedelta(minutes=delta_num)
        elif (duration_str == 'hour') or (duration_str == 'hours'):
            delta_num = int(time_strings_list[0])
            time_stamp = datetime.now() - timedelta(hours=delta_num)
        elif (duration_str == 'day') or (duration_str == 'days'):
            delta_num = int(time_strings_list[0])
            time_stamp = datetime.now() - timedelta(days=delta_num)
        elif (duration_str == 'month') or (duration_str == 'months'):
            # FOXNews mark Oct 14 as 1 month ago in a search result on Oct 30
            delta_num = int(time_strings_list[0]) * 30
            time_stamp = datetime.now() - timedelta(days=delta_num)
        else:
            date_format = '%b %d, %Y'
            time_stamp = datetime.strptime(time_string, date_format)
            
    except:
        time_stamp = parser.parse(time_string)
    
    return time_stamp
```

## NYTimes
https://www.nytimes.com/search?dropmab=false&endDate=2023-10-20&query=gaza&sort=newest&startDate=2023-10-01

```python
options = Options()
#options.headless = True
b = webdriver.Chrome(options=options)

url = 'https://www.nytimes.com/search?dropmab=false&endDate=2023-10-20&query=gaza&sort=newest&startDate=2023-10-01'
b.get(url)

time.sleep(1)
load_more_button = b.find_element(By.XPATH, '/html/body/div/div[2]/main/div/div[2]/div[3]/div/button')
```

```python
response_text = b.find_element(By.CLASS_NAME, 'css-8xl60i').get_attribute("outerHTML")

soup = BeautifulSoup(response_text, 'html.parser')

# Find the specific elements that contain the article links
div_list = soup.find_all('li', class_='css-1l4w6pd')

# for result in div_list:
#     article_url = result.find('a')['href']
#     print("Article URL:", article_url)

oldes_date_string = div_list[-1].find('span', class_='css-17ubb9w')['aria-label']
oldest_article_time = convert_to_local_time(oldes_date_string)
```

```python
div_list_length = 0

while oldest_article_time > datetime(2023, 10, 1, 0, 0):
    
    try:
        load_more_button.click()
        time.sleep(1)
    except:
        break
    
    response_text = b.find_element(By.CLASS_NAME, 'css-8xl60i').get_attribute("outerHTML")

    soup = BeautifulSoup(response_text, 'html.parser')

    # Find the specific elements that contain the article links
    div_list = soup.find_all('li', class_='css-1l4w6pd')
    
    # if the load more button doesn't return more reults, exit the loop
    if len(div_list) > div_list_length:
        div_list_length = len(div_list)
    else:
        break

    # for result in div_list:
    #     article_url = result.find('a')['href']
    #     print("Article URL:", article_url)

    oldes_date_string = div_list[-1].find('span', class_='css-17ubb9w')['aria-label']
    oldest_article_time = convert_to_local_time(oldes_date_string)
    
    time.sleep(1)
```

## WSJ
https://www.wsj.com/search?query=gaza&isToggleOn=true&operator=OR&sort=date-desc&duration=1y&source=wsjie

```python
options = Options()
#options.headless = True
b = webdriver.Chrome(options=options)

url = 'https://www.wsj.com/search?query=gaza&isToggleOn=true&operator=OR&sort=date-desc&duration=1y&source=wsjie'
b.get(url)
```

```python
# response_text = b.find_element(By.CLASS_NAME, 'WSJTheme--padding-top-large--2peQiWLU').get_attribute("outerHTML")
response_text = b.find_element(By.CLASS_NAME, 'style--column-8--2_beVGlu').get_attribute("outerHTML")

soup = BeautifulSoup(response_text, 'html.parser')

# Find the specific elements that contain the article links
div_list = soup.find_all('div', class_='WSJTheme--search-result--2NFlrTX7')
# div_list = div_list[:-1]

oldes_date_string = div_list[-1].find('div', class_='WSJTheme--timestamp--2zjbypGD').text
oldest_article_time = convert_to_local_time(oldes_date_string)
```

```python
oldest_article_time
```

```python
date_list = []
url_list = []
title_list = []

while oldest_article_time > datetime(2023, 10, 1, 0, 0):
    
    response_text = b.find_element(By.CLASS_NAME, 'style--column-8--2_beVGlu').get_attribute("outerHTML")
    soup = BeautifulSoup(response_text, 'html.parser')

    # Find the specific elements that contain the article links
    div_list = soup.find_all('div', class_='WSJTheme--search-result--2NFlrTX7')

    for result in div_list:
        date_string = result.find('div', class_='WSJTheme--timestamp--2zjbypGD').text
        article_date = convert_to_local_time(date_string)
        date_list.append(article_date)

        article_url = result.find('a', class_='')['href']
        url_list.append(article_url)

        article_title = result.find('span', class_='WSJTheme--headlineText--He1ANr9C').text
        title_list.append(article_title)
    
    
    # the last article date is the oldest
    oldest_article_time = article_date

    # click the next page link
    next_page_button = b.find_element(By.XPATH, '/html/body/div[1]/div/div/div/div[2]/div/div/div[2]/div[1]/div[5]/div/a[2]/span')
    next_page_button.click()
    time.sleep(2)


d = {'datetime': date_list, 'url': url_list, 'title':title_list}
df_WSJ_articles = pd.DataFrame(data=d)
```

```python
options = Options()
#options.headless = True
b = webdriver.Chrome(options=options)

url = 'https://www.reuters.com/search/news?blob=gaza&sortBy=date&dateRange=all'
b.get(url)

load_more_button = b.find_element(By.CLASS_NAME, 'search-result-more-txt')
```

```python
# b.find_element(By.CLASS_NAME, 'search-result-list').get_attribute("outerHTML")
```

```python
response_text = b.find_element(By.CLASS_NAME, 'search-result-list').get_attribute("outerHTML")

soup = BeautifulSoup(response_text, 'html.parser')

# Find the specific elements that contain the article links
div_list = soup.find_all('div', class_='search-result-indiv')

# for result in div_list:
#     article_url = result.find('a')['href']
#     print("Article URL:", article_url)

oldes_time_string = div_list[-1].find('h5').text
oldest_article_time = convert_to_local_time(oldes_time_string)
```

```python
div_list_length = 0

while oldest_article_time > datetime(2023, 10, 1, 0, 0):
    load_more_button.click()
    time.sleep(2)
    
    response_text = b.find_element(By.CLASS_NAME, 'search-result-list').get_attribute("outerHTML")

    soup = BeautifulSoup(response_text, 'html.parser')

    # Find the specific elements that contain the article links
    div_list = soup.find_all('div', class_='search-result-indiv')
    
    
    # if the load more button doesn't return more reults, exit the loop
    if len(div_list) > div_list_length:
        div_list_length = len(div_list)
    else:
        break
        
    # for result in div_list:
    #     article_url = result.find('a')['href']
    #     print("Article URL:", article_url)

    oldest_time_string = div_list[-1].find('h5').text
    oldest_article_time = convert_to_local_time(oldest_time_string)
```
